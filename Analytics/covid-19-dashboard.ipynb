```python
state_correction_dict = {
  'Bihar****': 'Bihar',
  'Dadra and Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
  'Madhya Pradesh***': 'Madhya Pradesh',
  'Maharashtra***': 'Maharashtra','Himachal Pradesh':'Himanchal Pradesh','Karnataka':'Karanataka',
'Telengana':'Telangana','Dadra and Nagar Haveli and Daman and Diu':'Daman & Diu'}
```

```python
def state_correction(state):
  try:
    return state_correction_dict[state]
  except:
    return state
```

```python
df_covid['State/UnionTerritory'] = df_covid['State/UnionTerritory'].apply(state_correction)
```

```python
df_covid['Active_cases'] = df_covid['Confirmed'] - (df_covid['Cured'] + df_covid['Deaths'])
```

```python
statewise = pd.pivot_table(df_covid, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
statewise
```

```python
statewise["Recovery_rate"] = statewise["Cured"]*100/statewise["Confirmed"]
statewise["Mortality_rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
statewise.sort_values(by = "Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```
